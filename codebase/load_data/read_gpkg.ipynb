```python
def read_layer_by_id(id):
    """
    Reads a specific layer from a GeoPackage file based on the provided identifier.
    
    Args:
        id (str): The identifier corresponding to the desired layer.
    
    Returns:
        GeoDataFrame: The GeoDataFrame containing the data from the specified layer.
    """
    if id not in ids:
        raise ValueError(f"Invalid ID: {id}. Must be one of {ids}")
    
    # Construct the file path
    file_path = os.path.join("data", "02_punt_tot_punt_analyse", f"{id}.gpkg")
    
    # Check if the file exists
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"GeoPackage file '{file_path}' not found.")
    
    # Read the layer into a GeoDataFrame
    gdf = gpd.read_file(file_path, layer=id)
    return gdf
```

```python
import os
import geopandas as gpd

# List available layers

#parameter for the geopandas read_file function
ids= ['basisscholen', 'bushaltes', 'buurt_centroiden', 'hbo_wo', 'highschools', 'ibis_centroiden', 'mbo', 'trainstations']
id = 'hbo_wo'


read_layer_by_id(id)
```
